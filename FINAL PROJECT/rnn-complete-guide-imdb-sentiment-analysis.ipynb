<div class="section-header"><span class="step-tag">01</span>Understanding Recurrent Neural Networks</div>

<div class="subsection-header">1.1 — Why Do We Need RNNs?</div>

<p>Traditional feedforward neural networks (and CNNs) assume that inputs are <b>independent</b> of one another.
But many real-world data types are inherently <b>sequential</b>, where order matters and earlier elements
influence the meaning of later ones:</p>

<div class="stat-grid">
<div class="stat-card"><div class="stat-number">📝</div><div class="stat-label">Text & Language</div></div>
<div class="stat-card"><div class="stat-number">🎙️</div><div class="stat-label">Speech & Audio</div></div>
<div class="stat-card"><div class="stat-number">📈</div><div class="stat-label">Time Series</div></div>
<div class="stat-card"><div class="stat-number">🧬</div><div class="stat-label">DNA Sequences</div></div>
</div>

<div class="callout callout-info">
<strong class="callout-title">💡 Core Idea</strong>
<p>A <b>Recurrent Neural Network (RNN)</b> introduces a <b>hidden state (memory)</b> that is updated at every
time step. The same weights are reused (shared) across all time steps, allowing the network to process
sequences of arbitrary length while "remembering" information from previous steps.</p>
</div>

<div class="subsection-header">1.2 — The Recurrence Formula</div>

<div class="callout callout-math">
<strong class="callout-title">🧮 Mathematical Formulation</strong>
<p>At each time step <code>t</code>, the RNN cell computes a new hidden state <code>h_t</code> from the
current input <code>x_t</code> and the previous hidden state <code>h_(t-1)</code>:</p>
<p style="text-align:center; font-size:1.15em;">
<code>h_t = tanh(W_xh · x_t + W_hh · h_(t-1) + b_h)</code><br>
<code>y_t = W_hy · h_t + b_y</code>
</p>
<p>Where <code>W_xh</code>, <code>W_hh</code>, <code>W_hy</code> are weight matrices shared across <b>every</b>
time step, and <code>b_h</code>, <code>b_y</code> are bias terms. This weight-sharing is what gives RNNs the
ability to generalize across sequences of different lengths.</p>
</div>

<div class="subsection-header">1.3 — Unrolling the RNN Through Time</div>

<p>Conceptually, an RNN can be "unrolled" across time steps, turning the recurrent loop into a chain of
identical cells, each passing its hidden state forward:</p>

<div class="flow-steps">
<div class="flow-step">x₁ → h₁</div><div class="flow-arrow">→</div>
<div class="flow-step">x₂, h₁ → h₂</div><div class="flow-arrow">→</div>
<div class="flow-step">x₃, h₂ → h₃</div><div class="flow-arrow">→</div>
<div class="flow-step">... → h_T</div><div class="flow-arrow">→</div>
<div class="flow-step">Output / Prediction</div>
</div>

<div class="callout callout-warning">
<strong class="callout-title">⚠️ The Vanishing / Exploding Gradient Problem</strong>
<p>Because <b>Backpropagation Through Time (BPTT)</b> multiplies gradients across every time step, gradients
can shrink toward zero (<b>vanishing</b>) or grow uncontrollably (<b>exploding</b>) as sequences get longer.
This makes it very difficult for vanilla (Simple) RNNs to learn <b>long-range dependencies</b> — e.g.
connecting a word at the start of a paragraph to a sentiment expressed at the end.</p>
</div>



<div class="section-header"><span class="step-tag">02</span>Solving the Memory Problem: LSTM & GRU</div>

<div class="subsection-header">2.1 — Long Short-Term Memory (LSTM)</div>

<p>LSTMs (Hochreiter & Schmidhuber, 1997) introduce a separate <b>cell state (C_t)</b> that acts as a
"conveyor belt" of memory, controlled by three <b>gates</b> that learn what to keep, forget, and output:</p>

<div class="comparison-grid">
<div class="comparison-card">
<div class="comparison-card-header">🚪 Forget Gate</div>
<div class="comparison-card-body">
<code>f_t = σ(W_f·[h_(t-1), x_t] + b_f)</code>
<ul><li>Decides what fraction of the old cell state to discard</li>
<li>Output near 0 → forget, near 1 → retain</li></ul>
</div>
</div>
<div class="comparison-card">
<div class="comparison-card-header alt">📥 Input Gate</div>
<div class="comparison-card-body">
<code>i_t = σ(W_i·[h_(t-1), x_t] + b_i)</code><br>
<code>C̃_t = tanh(W_c·[h_(t-1), x_t] + b_c)</code>
<ul><li>Decides how much new information to add</li>
<li>Updates cell state: <code>C_t = f_t * C_(t-1) + i_t * C̃_t</code></li></ul>
</div>
</div>
<div class="comparison-card">
<div class="comparison-card-header">📤 Output Gate</div>
<div class="comparison-card-body">
<code>o_t = σ(W_o·[h_(t-1), x_t] + b_o)</code><br>
<code>h_t = o_t * tanh(C_t)</code>
<ul><li>Controls what part of the cell state becomes the visible hidden state</li></ul>
</div>
</div>
</div>

<div class="callout callout-success">
<strong class="callout-title">✅ Why LSTM Works Better</strong>
<p>The cell state <code>C_t</code> has an (almost) uninterrupted gradient path across time steps
(additive updates rather than repeated multiplications), which dramatically reduces the vanishing
gradient problem and allows LSTMs to retain information over hundreds of time steps.</p>
</div>

<div class="subsection-header">2.2 — Gated Recurrent Unit (GRU)</div>

<p>GRUs (Cho et al., 2014) simplify the LSTM by merging the cell state and hidden state, and combining
the forget/input gates into a single <b>update gate</b>:</p>

<div class="callout callout-math">
<strong class="callout-title">🧮 GRU Equations</strong>
<p>
<code>z_t = σ(W_z·[h_(t-1), x_t])</code> &nbsp;(update gate)<br>
<code>r_t = σ(W_r·[h_(t-1), x_t])</code> &nbsp;(reset gate)<br>
<code>h̃_t = tanh(W·[r_t * h_(t-1), x_t])</code><br>
<code>h_t = (1 - z_t) * h_(t-1) + z_t * h̃_t</code>
</p>
</div>

<div class="subsection-header">2.3 — Bidirectional RNNs</div>

<div class="callout callout-tip">
<strong class="callout-title">🔄 Reading in Both Directions</strong>
<p>A <b>Bidirectional RNN/LSTM</b> processes the sequence both <b>left-to-right</b> and <b>right-to-left</b>
using two separate recurrent layers, then concatenates their hidden states. For sentiment analysis,
this lets the model use context from <i>both</i> the beginning and end of a review when interpreting
any given word — often boosting accuracy at the cost of roughly doubled computation.</p>
</div>

<div class="subsection-header">2.4 — Architecture Cheat-Sheet</div>

<div class="comparison-grid">
<div class="comparison-card">
<div class="comparison-card-header">SimpleRNN</div>
<div class="comparison-card-body">
<ul>
<li>Single tanh activation per cell</li>
<li>Fewest parameters, fastest to train</li>
<li>Struggles with long sequences (vanishing gradients)</li>
</ul>
<div class="verdict">Best for: short sequences, baselines</div>
</div>
</div>
<div class="comparison-card">
<div class="comparison-card-header alt">LSTM</div>
<div class="comparison-card-body">
<ul>
<li>3 gates + cell state (4x parameters of SimpleRNN)</li>
<li>Excellent long-range memory</li>
<li>Slower to train, more memory-intensive</li>
</ul>
<div class="verdict">Best for: long sequences, complex dependencies</div>
</div>
</div>
<div class="comparison-card">
<div class="comparison-card-header">GRU</div>
<div class="comparison-card-body">
<ul>
<li>2 gates, no separate cell state (3x parameters of SimpleRNN)</li>
<li>Comparable accuracy to LSTM, fewer parameters</li>
<li>Often trains faster than LSTM</li>
</ul>
<div class="verdict">Best for: efficient long-range modeling</div>
</div>
</div>
<div class="comparison-card">
<div class="comparison-card-header alt">Bidirectional LSTM</div>
<div class="comparison-card-body">
<ul>
<li>Two LSTMs (forward + backward) concatenated</li>
<li>Captures past <i>and</i> future context</li>
<li>~2x parameters & compute of unidirectional LSTM</li>
</ul>
<div class="verdict">Best for: tasks where full-sequence context helps</div>
</div>
</div>
</div>


<div class="subsection-header">4.2 — Exploratory Data Analysis</div>
<p>Let's inspect the class balance, review-length distribution, and decode a sample review back into
human-readable text using the IMDB word index.</p>


```python
word_index = imdb.get_word_index()
reverse_index = {v + 3: k for k, v in word_index.items()}
reverse_index[0] = "<PAD>"
reverse_index[1] = "<START>"
reverse_index[2] = "<UNK>"
reverse_index[3] = "<UNUSED>"

def decode_review(encoded_review):
    return " ".join(reverse_index.get(i, "?") for i in encoded_review)

print("Decoded sample review:\n")
print(decode_review(x_train_raw[0][:60]), "...")
print("\nSentiment label:", "Positive 👍" if y_train[0] == 1 else "Negative 👎")

unique, counts = np.unique(y_train, return_counts=True)
print("\nTraining class balance:", dict(zip(unique, counts)))

lengths = [len(r) for r in x_train_raw]
print(f"Min length: {min(lengths)} | Max length: {max(lengths)} | "
      f"Mean length: {np.mean(lengths):.1f} | Median length: {np.median(lengths):.0f}")
```


<div class="section-header"><span class="step-tag">05</span>Preprocessing: Padding the Sequences</div>

<div class="callout callout-tip">
<strong class="callout-title">📏 Why Padding is Required</strong>
<p>RNN layers in Keras process batches as fixed-size tensors. Since reviews have variable lengths,
we use <code>pad_sequences</code> to make every review exactly <code>MAX_LEN = 200</code> tokens —
truncating long reviews and zero-padding short ones at the <b>start</b> of the sequence
(<code>padding="pre"</code>), which empirically tends to work better for RNNs since the most
recent (and most informative) tokens stay closer to the final hidden state.</p>
</div>


```python
x_train = pad_sequences(x_train_raw, maxlen=MAX_LEN, padding="pre", truncating="pre")
x_test  = pad_sequences(x_test_raw,  maxlen=MAX_LEN, padding="pre", truncating="pre")

print("x_train shape:", x_train.shape)
print("x_test shape :", x_test.shape)
print("\nExample padded sequence (first 30 tokens):")
print(x_train[0][:30])
```


<div class="subsection-header">5.1 — Shared Configuration & Helper Functions</div>
<p>To keep comparisons fair, every model below shares the same embedding size, batch size, number
of epochs, and evaluation/plotting helper functions.</p>


```python
EMBEDDING_DIM = 32
BATCH_SIZE = 128
EPOCHS = 5

history_store = {}  
results_store = {}  

def compile_and_train(model, name, x_tr, y_tr, x_val, y_val):
    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    print(model.summary())
    history = model.fit(
        x_tr, y_tr,
        validation_data=(x_val, y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        verbose=1
    )
    history_store[name] = history.history
    return history

def evaluate_model(model, name, x_te, y_te):
    loss, acc = model.evaluate(x_te, y_te, verbose=0)
    y_pred_proba = model.predict(x_te, verbose=0).ravel()
    y_pred = (y_pred_proba > 0.5).astype(int)
    results_store[name] = {
        "loss": loss, "accuracy": acc,
        "y_pred": y_pred, "y_pred_proba": y_pred_proba
    }
    print(f"[{name}] Test Loss: {loss:.4f} | Test Accuracy: {acc:.4f}")
    return loss, acc

def plot_training_curves(name):
    h = history_store[name]
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    axes[0].plot(h["accuracy"], color=THEME["olive"], marker="o", label="Train")
    axes[0].plot(h["val_accuracy"], color=THEME["brown"], marker="o", label="Validation")
    axes[0].set_title(f"{name}: Accuracy per Epoch", fontweight="bold")
    axes[0].set_xlabel("Epoch"); axes[0].set_ylabel("Accuracy"); axes[0].legend()

    axes[1].plot(h["loss"], color=THEME["olive"], marker="o", label="Train")
    axes[1].plot(h["val_loss"], color=THEME["brown"], marker="o", label="Validation")
    axes[1].set_title(f"{name}: Loss per Epoch", fontweight="bold")
    axes[1].set_xlabel("Epoch"); axes[1].set_ylabel("Loss"); axes[1].legend()

    plt.tight_layout()
    plt.show()
```


<div class="section-header"><span class="step-tag">06</span>Model 1 — SimpleRNN Baseline</div>

<div class="callout callout-info">
<strong class="callout-title">🏗️ Architecture</strong>
<p>An <code>Embedding</code> layer converts integer tokens into dense 32-dimensional vectors,
followed by a single <code>SimpleRNN</code> layer with 32 units, then a <code>Dense</code>
sigmoid output for binary classification.</p>
</div>


```python
model_rnn = Sequential([
    Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM, input_length=MAX_LEN),
    SimpleRNN(32, activation="tanh"),
    Dropout(0.3),
    Dense(1, activation="sigmoid")
], name="SimpleRNN_Model")

history_rnn = compile_and_train(model_rnn, "SimpleRNN", x_train, y_train, x_test, y_test)
```

```python
evaluate_model(model_rnn, "SimpleRNN", x_test, y_test)
plot_training_curves("SimpleRNN")
```


<div class="section-header"><span class="step-tag">07</span>Model 2 — LSTM (Long Short-Term Memory)</div>

<div class="callout callout-success">
<strong class="callout-title">🏗️ Architecture</strong>
<p>Replacing <code>SimpleRNN</code> with <code>LSTM</code> (same 32 units) gives the model
gated memory via the forget/input/output gates discussed in Section 2 — expect noticeably
better handling of long reviews.</p>
</div>


```python
model_lstm = Sequential([
    Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM, input_length=MAX_LEN),
    LSTM(32, activation="tanh"),
    Dropout(0.3),
    Dense(1, activation="sigmoid")
], name="LSTM_Model")

history_lstm = compile_and_train(model_lstm, "LSTM", x_train, y_train, x_test, y_test)
```

```python
evaluate_model(model_lstm, "LSTM", x_test, y_test)
plot_training_curves("LSTM")
```


<div class="section-header"><span class="step-tag">08</span>Model 3 — GRU (Gated Recurrent Unit)</div>

<div class="callout callout-success">
<strong class="callout-title">🏗️ Architecture</strong>
<p>The <code>GRU</code> layer uses fewer gates than LSTM (update + reset vs. forget + input + output),
typically giving comparable accuracy with fewer parameters and faster training per epoch.</p>
</div>


```python
model_gru = Sequential([
    Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM, input_length=MAX_LEN),
    GRU(32, activation="tanh"),
    Dropout(0.3),
    Dense(1, activation="sigmoid")
], name="GRU_Model")

history_gru = compile_and_train(model_gru, "GRU", x_train, y_train, x_test, y_test)
```

```python
evaluate_model(model_gru, "GRU", x_test, y_test)
plot_training_curves("GRU")
```


<div class="section-header"><span class="step-tag">09</span>Model 4 — Bidirectional LSTM</div>

<div class="callout callout-success">
<strong class="callout-title">🏗️ Architecture</strong>
<p>Wrapping the LSTM layer in <code>Bidirectional</code> runs two LSTM passes — one forward, one
backward over the review — and concatenates their final hidden states (64 total units) before the
output layer. This typically yields the best accuracy among the four models, at roughly double the
training time.</p>
</div>


```python
model_bilstm = Sequential([
    Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM, input_length=MAX_LEN),
    Bidirectional(LSTM(32, activation="tanh")),
    Dropout(0.3),
    Dense(1, activation="sigmoid")
], name="BiLSTM_Model")

history_bilstm = compile_and_train(model_bilstm, "BiLSTM", x_train, y_train, x_test, y_test)
```

```python
evaluate_model(model_bilstm, "BiLSTM", x_test, y_test)
plot_training_curves("BiLSTM")
```


<div class="section-header"><span class="step-tag">11</span>Visualizing the Learned Word Embeddings</div>

<div class="callout callout-info">
<strong class="callout-title">🔍 What Are We Looking At?</strong>
<p>Every RNN above starts with an <code>Embedding</code> layer that learns a 32-dimensional vector
for each of the 10,000 vocabulary words <i>during sentiment training</i>. Words that play similar
roles for sentiment classification should end up with similar embedding vectors. We extract the
embedding weight matrix from the <b>LSTM model</b> and project it down to 2D using <b>t-SNE</b>
for visualization.</p>
</div>

<div class="callout callout-tip">
<strong class="callout-title">🛠️ Extraction Note</strong>
<p>For <code>Sequential</code> models, the embedding weight matrix can be read directly from the
trained <code>Embedding</code> layer via <code>layer.get_weights()[0]</code> — no need to build an
intermediate <code>keras.Model</code>, which avoids the <code>AttributeError</code> that occurs when
constructing intermediate models from layers of an already-built <code>Sequential</code> model.</p>
</div>


```python
embedding_layer = model_lstm.layers[0]
embedding_weights = embedding_layer.get_weights()[0]
print("Embedding matrix shape:", embedding_weights.shape)

inv_word_index = {v: k for k, v in word_index.items() if v < VOCAB_SIZE}

positive_words = ["great", "excellent", "wonderful", "amazing", "love", "best",
                   "fantastic", "brilliant", "perfect", "enjoyed"]
negative_words = ["terrible", "awful", "worst", "boring", "horrible", "bad",
                   "disappointing", "waste", "poor", "dull"]
neutral_words   = ["movie", "film", "actor", "scene", "director", "story",
                    "character", "plot", "time", "watch"]

all_words = positive_words + negative_words + neutral_words
word_to_idx = {w: word_index[w] + 3 for w in all_words if w in word_index and (word_index[w] + 3) < VOCAB_SIZE}

vectors = np.array([embedding_weights[idx] for idx in word_to_idx.values()])
labels = list(word_to_idx.keys())

print(f"Resolved {len(labels)} of {len(all_words)} curated words within the vocabulary.")
```

```python
tsne = TSNE(n_components=2, perplexity=8, random_state=42, init="pca")
vectors_2d = tsne.fit_transform(vectors)

plt.figure(figsize=(9, 7))
for i, word in enumerate(labels):
    if word in positive_words:
        color = THEME["olive"]
    elif word in negative_words:
        color = THEME["brown"]
    else:
        color = THEME["orange"]
    plt.scatter(vectors_2d[i, 0], vectors_2d[i, 1], color=color, s=80, edgecolor=THEME["bg"])
    plt.annotate(word, (vectors_2d[i, 0], vectors_2d[i, 1]),
                  textcoords="offset points", xytext=(6, 4), fontsize=10, color=THEME["bg"])

from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Positive words', markerfacecolor=THEME["olive"], markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Negative words', markerfacecolor=THEME["brown"], markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Neutral words', markerfacecolor=THEME["orange"], markersize=10),
]
plt.legend(handles=legend_elements, loc="best")
plt.title("t-SNE Projection of LSTM-Learned Word Embeddings", fontweight="bold")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.tight_layout()
plt.show()
```


<div class="section-header"><span class="step-tag">12</span>Try It Yourself: Predict Custom Reviews</div>

<div class="callout callout-tip">
<strong class="callout-title">✍️ Live Inference</strong>
<p>Write your own short movie review below and see how each trained model scores its sentiment
(closer to <b>1.0</b> = positive, closer to <b>0.0</b> = negative).</p>
</div>


```python
def encode_review(text, vocab_size=VOCAB_SIZE, maxlen=MAX_LEN):
    tokens = text.lower().split()
    encoded = [1]  
    for word in tokens:
        idx = word_index.get(word, 2) + 3  
        if idx >= vocab_size:
            idx = 2
        encoded.append(idx)
    return pad_sequences([encoded], maxlen=maxlen, padding="pre", truncating="pre")

sample_reviews = [
    "this movie was absolutely wonderful the acting was great and i loved every minute",
    "what a terrible boring film i wasted my time watching this awful disaster",
    "the plot was okay but the acting felt a bit flat overall an average movie"
]

models_dict = {
    "SimpleRNN": model_rnn, "LSTM": model_lstm,
    "GRU": model_gru, "BiLSTM": model_bilstm
}

for review in sample_reviews:
    encoded = encode_review(review)
    print(f"\nReview: \"{review}\"")
    for name, model in models_dict.items():
        score = model.predict(encoded, verbose=0)[0][0]
        sentiment = "Positive 👍" if score > 0.5 else "Negative 👎"
        print(f"  {name:10s} -> score = {score:.3f}  ({sentiment})")
```
